```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
#Loading Data
iris=datasets.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.tail() #Checking
```

```python
#Graphs
sepal_width=iris['data'][:,1]
petal_width=iris['data'][:,3]
sepal_length=iris['data'][:,0]
petal_length=iris['data'][:,2]
c=iris['target']              
plt.scatter(sepal_width,petal_width,c=c)
plt.title('PETAL WIDTH VS SEPAL WIDTH')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.show()
plt.scatter(sepal_length,petal_length,c=c)
plt.title('PETAL LENGTH VS SEPAL LENGTH')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.show()
```

```python
#Splitting data into test and train sets
x=np.column_stack((sepal_width,petal_width,sepal_length,petal_length))
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
#Linear Model
model1=svm.SVC(kernel='linear')
model1.fit(x_train,y_train)
print('Accuracy of Linear Model is:',model1.score(x_test,y_test)*100,'%')
```

```python
#Polynomial Model
model2=svm.SVC(kernel='poly')
model2.fit(x_train,y_train)
print('Accuracy of Polynomial Model is:',model2.score(x_test,y_test)*100,'%')
```

```python
#Sigmoid Model
model3=svm.SVC(kernel='sigmoid')
model3.fit(x_train,y_train)
print('Accuracy of Sigmoid Model is:',model3.score(x_test,y_test)*100,'%')
```

```python
#RBF/Gaussian Model
model4=svm.SVC(kernel='rbf')
model4.fit(x_train,y_train)
print('Accuracy of RBF/Gaussian Model is:',model4.score(x_test,y_test)*100,'%')
```
